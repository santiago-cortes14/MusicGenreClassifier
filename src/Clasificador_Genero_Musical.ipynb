La base de datos consta de **1.000 muestras**, cada una con **26 características** del dominio temporal y frecuencial, obtenidas de **10 géneros musicales**: *Blues*, *Classical*, *Country*, *Disco*, *Hip Hop*, *Jazz*, *Metal*, *Pop*, *Reggae* & *Rock*.

```python
df = pd.read_csv('music_genre.csv')
print(df.columns)
```

Teniendo presente que, la base de datos presenta una uniformidad muestral de **100 muestras** en cada clase, se trabajó con **400 muestras**, pues se descartaron los géneros: *Blues*, *Country*, *Disco*, *Hip Hop*, *Jazz* y *Reggae*, debido a que no beneficiaban al parámetro de evaluación.

```python
print("Tamaño de la Base de Datos Inicial:",df.shape)

df = df[df.label != 0]
df = df[df.label != 2]
df = df[df.label != 3]
df = df[df.label != 4]
df = df[df.label != 5]
df = df[df.label != 8]

df.loc[df['label'] == 1,'label'] = 0
df.loc[df['label'] == 6,'label'] = 1
df.loc[df['label'] == 7,'label'] = 2
df.loc[df['label'] == 9,'label'] = 3
df = df.reset_index(drop = True)

nombre_clases = ['Classical', 'Metal', 'Pop', 'Rock']
n_clases = len(nombre_clases)

n_songs = len(df)

df = df.drop(columns = 'filename')
print("Tamaño de la Base de Datos Final:",df.shape)

target = np.zeros([n_songs,1])
target[:,0] = df['label']
print("Tamaño de la Matriz Target:",target.shape)

df = df.drop(columns = 'label')

n_descriptores = len(df.columns)

nombre_descriptores = []
for f in range(n_descriptores):
    nombre_descriptores.append(df.columns[f])

descriptores = np.zeros([n_songs,n_descriptores])
for d in range(n_descriptores):
    descriptores[:,d] = df.iloc[:,d]
print("Tamaño de la Matriz Features:",descriptores.shape)

pos_clase = []
for x in range(n_clases):
    pos_clase.append(np.where(target[:,0] == x))
```

Seguidamente, se analizaron los descriptores para descartar los aportaran positivamente a la clasificación. Para esto, se asumió que los datos presentaban una distribución Gaussiana.

```python
def p_normal(x, m, sd):
    p_normal = 1/(sd*np.sqrt(np.pi*2))*np.exp(-0.5*np.power((x-m)/sd,2))
    return p_normal
```

Al observar la escasa separabilidad de los descriptores, se implementó el Análisis de Componentes Principales (PCA) en búsqueda de mejorar el criterio de evaluación reduciendo las componentes. Sin embargo, al no ver mejoría, se optó por trabajar con los 26 descriptores, confiando en que algunos de estos realizarían un aporte en la clasificación específica de ciertas clases.

```python
for i in range(n_descriptores):
    descriptor = descriptores[:,i]

    m_estad_clase_descriptor = np.zeros([2,n_clases])
    for y in range(n_clases):
        m_estad_clase_descriptor[0,y] = np.mean(descriptor[pos_clase[y]])
        m_estad_clase_descriptor[1,y] = np.std(descriptor[pos_clase[y]])

    eje_x = np.linspace(np.min(descriptores[:,i]),np.max(descriptores[:,i]),100)

    F_clase = np.zeros([len(eje_x),n_clases])
    for y in range(n_clases):
        F_clase[:,y] = p_normal(eje_x, m_estad_clase_descriptor[0,y], m_estad_clase_descriptor[1,y])

    plt.figure()
    plt.title("Gaussian Distribution of Descriptor " + str(i))
    for y in range(n_clases):
        plt.plot(eje_x, F_clase[:,y], label = nombre_clases[y])
    plt.xlabel(nombre_descriptores[i])
    plt.ylabel("frequency")
    plt.legend()
    plt.grid()
    plt.show()

descriptores_clf = range(26)

n_descriptores_clf = len(descriptores_clf)

x = np.zeros([n_songs,n_descriptores_clf])
for i in range(n_descriptores_clf):
    x[:,i] = descriptores[:,descriptores_clf[i]]

y = np.zeros([n_songs,1])
for j in range(n_clases):
    y[pos_clase[j]] = j
```

Una vez se definieron los descriptores, se dividieron las muestras en **280** para **entrenamiento** (70%) y **120** para **validación** (30%).

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

En esta ocasión, se entrenó el clasificador una única vez y se exportó el modelo en un archivo *(classifier.joblib)*. De esta forma, se evitó la repetición del proceso de entrenamiento, y a su vez ligeros cambios en el modelo debido a variaciones en la convergencia durante el entrenamiento.

```python
# clf = GradientBoostingClassifier()

# clf.fit(x_train,y_train)

# jl.dump(clf, 'classifier.joblib')
```

Despues, se importó el modelo de clasificación seleccionado.

```python
clf = jl.load('classifier.joblib')
```

Luego de importar el modelo, se realizó la predicción con los datos de validación.

```python
y_pred = clf.predict(x_test)
```

Finalmente, se evaluó el desempeño del clasificador a partir del F1-Score.

```python
f1 = f1_score(y_test, y_pred, average = None)

f1_total = f1_score(y_test, y_pred, average = 'macro')

print("\n------------------------- F1-SCORE -------------------------")

print("\n---------- Gradient Boosting ----------")
for c in range(n_clases):
    print("Clase " + nombre_clases[c] + ": " + str(round(f1[c], 2)))
print("Total:",round(f1_total, 4))
```

## **IMPLEMENTACIÓN Y RESULTADOS DE TODAS LAS METODOLOGÍAS**

La necesidad de plantear diversas metodologías se presentó por el criterio de evaluación, pues inicialmente su valor fue regular. No osbtante, se observó que esta era bajo para clases específicas y al ser promediado con criterios altos, afectaba negativamente el total.

En cada metodología se conservaron los modelos que destacaron en mínimo una de la descomposiciones ejecutadas (incluyendo donde no se descompuso). Las técnicas que se mantuvieron entre las metodologías fueron:

> Bosques Aleatorios (RandomForestClassifier).

> Árboles Extremadamente Aleatorios (ExtraTreesClassifier).

> Centroide Más Cercano (NearestCentroid).

> Naive Bayes Modelo Gaussiano (GaussianNB).

> Perceptrón MultiCapa (MLPClassifier).

> Soporte Vectorial Lineal (LinearSVC).

```python
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.neighbors import NearestCentroid
from sklearn.naive_bayes import  GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.svm import LinearSVC
```

El PCA ofrecía la posibilidad de descomponer los descriptores en mínimo 26 componentes y máximo una componente. Por tanto, en las metodologías donde se aplicó, se modificó una variable ***componentes*** desde 1 hasta 26.

### **Sin Excluir Clases**

```python
df = pd.read_csv('music_genre.csv')

nombre_clases = ['Blues', 'Classical', 'Country', 'Disco', 'Hip Hop', 'Jazz', 'Metal', 'Pop', 'Reggae', 'Rock']
n_clases = len(nombre_clases)

n_songs = len(df)

target = np.zeros([n_songs,1])
target[:,0] = df['label']

df = df.drop(columns = 'filename')
df = df.drop(columns = 'label')

n_descriptores = len(df.columns)

nombre_descriptores = []
for f in range(n_descriptores):
    nombre_descriptores.append(df.columns[f])

descriptores = np.zeros([n_songs,n_descriptores])
for d in range(n_descriptores):
    descriptores[:,d] = df.iloc[:,d]

pos_clase = []
for x in range(n_clases):
    pos_clase.append(np.where(target[:,0] == x))

descriptores_clf = range(26)

n_descriptores_clf = len(descriptores_clf)

x = np.zeros([n_songs,n_descriptores_clf])
for i in range(n_descriptores_clf):
    x[:,i] = descriptores[:,descriptores_clf[i]]

y = np.zeros([n_songs,1])
for j in range(n_clases):
    y[pos_clase[j]] = j

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)

clf_rf = RandomForestClassifier(n_estimators = 50, max_depth = 10, random_state = 0)
clf_nc = NearestCentroid()
clf_et = ExtraTreesClassifier()

clf_rf.fit(x_train,y_train)
clf_nc.fit(x_train,y_train)
clf_et.fit(x_train,y_train)

y_pred_rf = clf_rf.predict(x_test)
y_pred_nc = clf_nc.predict(x_test)
y_pred_et = clf_et.predict(x_test)

f1_rf = f1_score(y_test, y_pred_rf, average = None)
f1_nc = f1_score(y_test, y_pred_nc, average = None)
f1_et = f1_score(y_test, y_pred_et, average = None)

f1_rf_total = f1_score(y_test, y_pred_rf, average = 'macro')
f1_nc_total = f1_score(y_test, y_pred_nc, average = 'macro')
f1_et_total = f1_score(y_test, y_pred_et, average = 'macro')

print("\n------------------------- F1-SCORE -------------------------")

print("\n---------- Random Forest ----------")
for c in range(n_clases):
    print("Clase " + nombre_clases[c] + ": " + str(round(f1_rf[c], 2)))
print("Total:",round(f1_rf_total, 4))

print("\n---------- Nearest Centroid ----------")
for c in range(n_clases):
    print("Clase " + nombre_clases[c] + ": " + str(round(f1_nc[c], 2)))
print("Total:",round(f1_nc_total, 4))

print("\n---------- Extra Trees ----------")
for c in range(n_clases):
    print("Clase " + nombre_clases[c] + ": " + str(round(f1_et[c], 2)))
print("Total:",round(f1_et_total, 4))
```
